```python
# Your code here
# Your code here

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Division by 0 error"

def power(x, y):
    return x ** y

def square_root(n):
    return n ** 0.5

def mod(x, y):
    if y != 0:
        return x % y
    else:
        return "Modulus by 0 error"

def compute(num1, num2, operator):
    if operator == '+':
        return add(num1, num2)
    elif operator == '-':
        return subtract(num1, num2)
    elif operator == '*':
        return multiply(num1, num2)
    elif operator == '/':
        return divide(num1, num2)
    elif operator == '**':
        return power(num1, num2)
    elif operator == '%':
        return mod(num1, num2)
    else:
        return "Invalid operator"

def simple_calculator():
    print("Supported operations: +, -, *, /, ** (power), %, sqrt (square root)")
    while True:
        try:
            value1 = float(input("Enter the first number: "))
            op = input("Enter operator (or 'sqrt' for square root): ")
            if op == 'sqrt':
                print(f"‚àö{value1} = {square_root(value1):.2f}")
            else:
                value2 = float(input("Enter the second number: "))
                result = compute(value1, value2, op)
                print(f"{value1} {op} {value2} = {result}")
        except ValueError:
            print("Invalid input. Please enter numeric values.")

        exit_choice = input("exit? (y/n): ").lower()
        if exit_choice == 'y':
            break

if __name__ == "__main__":
    simple_calculator()
```

# Question 5 (20 points).

In the field of Data Science, data is often formatted as a comma-delimited (CSV) file, where each line in a file corresponds to a field's value. Refer to Chapter 9 (p. 106) for more information on comma-delimited files.

To answer this question, you will need to download the file [Assignment1_denton_housing.csv](https://raw.githubusercontent.com/unt-iialab/info5731-spring2022/main/assignments/Assignment1_denton_housing.csv) included with this assignment. The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----
	2012 ----- -----
    2013 ----- -----
	2014 ----- -----
  
  

(c) (5 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (5 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
import pandas as pd
import numpy as np

# Read the CSV file
df = pd.read_csv('Assignment1_denton_housing.csv')

# Creating a pivot table for analysis
df_pivot = df.pivot(index='year', columns='title_field', values='value')

# (a) Calculate year-year differences in Occupied Housing Units
print("(a) Year-over-year differences in Occupied Housing Units:")
differences = []
years = sorted(df_pivot.index.unique())
for i in range(len(years)-1):
    curr_year = df_pivot.loc[years[i], 'Occupied Housing Units']
    next_year = df_pivot.loc[years[i+1], 'Occupied Housing Units']
    diff = next_year - curr_year
    differences.append(diff)
    print(f"{years[i]}-{years[i+1]}: {diff}")

print("\nDifferences in ascending order:", sorted(differences))

# (b) Calculating the percentage of vacant and occupied units
print("\n(b) Housing unit percentages:")
print("Year\tVacant Housing Units\tOccupied Housing Units")
for year in years:
    total = df_pivot.loc[year, 'Total Housing Units']
    vacant_pct = (df_pivot.loc[year, 'Vacant Housing Units'] / total) * 100
    occupied_pct = (df_pivot.loc[year, 'Occupied Housing Units'] / total) * 100
    print(f"{year}\t{vacant_pct:.1f}%\t\t\t{occupied_pct:.1f}%")

# (c) Finding the years with highest vacant and occupied units
print("\n(c) Peak years:")
max_vacant_year = df_pivot['Vacant Housing Units'].idxmax()
max_occupied_year = df_pivot['Occupied Housing Units'].idxmax()

print("\t\t\tYear\tValue")
print(f"Vacant Housing Units\t{max_vacant_year}\t{df_pivot.loc[max_vacant_year, 'Vacant Housing Units']:.0f}")
print(f"Occupied Housing Units\t{max_occupied_year}\t{df_pivot.loc[max_occupied_year, 'Occupied Housing Units']:.0f}")

# (d) Calculating the harmonic mean of total housing units
total_units = df_pivot['Total Housing Units']
harmonic_mean = len(total_units) / sum(1/x for x in total_units)
print(f"\n(d) Harmonic mean of total housing units: {harmonic_mean:.2f}")
```
